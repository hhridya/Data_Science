In box plot we could see that there are outliers so we will remove outliers by replacing those value with upper whisker value

```python
def remove_outliers(colname):
  
  quantile_75 = application_data[colname].quantile(0.75) # 75% quantile value of boxplot i.e. Q3
  quantile_25 = application_data[colname].quantile(0.25) # 25%  quantile value of boxplot i.e. Q1
  upperlimit =  quantile_75 + (quantile_75 - quantile_25) * 1.5 # calculating upper limit = Q3+1.5(Q1+Q3)
  condition = application_data[colname] < upperlimit
  application_data[colname].where(condition,upperlimit,inplace=True)

for i in list_for_continuous:
  remove_outliers(i)
```

### Categorical variables

NAME_CONTRACT_TYPE v/s NAME_CONTRACT_STATUS

```python
CONTRACT_STATUS = 'NAME_CONTRACT_STATUS'
NCT = 'NAME_CONTRACT_TYPE'
CS = CONTRACT_STATUS
def compare_categorical(col1,col2):

  res = previous_application.groupby([col1,col2])[col2].count().unstack().plot.bar()

  plt.show()

compare_categorical(NCT,CS)
```

There are few categorical values in data like NAME_CONTRACT_TYPE, NAME_CLIENT_TYPE for which we need to see how these categorical value affects being defaulter or not.
From above graph we could see that proportion of XNA is very small in comparison to other NAME_CONTRACT_TYPE categories. So to normalize this uneveness will follow below steps:

* take count of category and target by grouping them 
* take total count of each category
* divide count by the total to get normalized proportions


```python
CONTRACT_STATUS = 'NAME_CONTRACT_STATUS'
NCT = 'NAME_CONTRACT_TYPE'
CS = CONTRACT_STATUS
def compare_categorical(col1,col2):

  res = previous_application.groupby([col1,col2])[col2].count().unstack()
  
  res_sum = res.sum(axis=1)  # sum along rows
  res_final = res.divide(res_sum,axis=0).fillna(0)*100  # calculate % after replacing null with 0
  res_final.plot.bar()

  plt.ylabel('Normalized_Count')
  plt.show()
```
